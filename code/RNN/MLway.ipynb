## 调用此函数获得线性模型的输出 array（可作为特征）

```python
def LinearResult(model,df):
    X_test=df.loc[:,['soc','SOCgap','temp_max','temp_min','tempMAXMINdiff']].to_numpy()
    y_test=df.loc[:,['milegap']].to_numpy()
    y_pred=model.predict(X_test)
    return y_pred,y_test
```

# 那么重新训练一个SGDRegressor
#### 创建迭代器：

```python
LinearModel_X = lambda df:df.loc[:,['soc','SOCgap','temp_max','temp_min','tempMAXMINdiff']].to_numpy()
LinearModel_y = lambda df:df.loc[:,['milegap']].to_numpy()
def iter_minibatches(dflist):
    '''
    迭代器
    给定文件流（比如一个大文件），每次输出minibatch_size行，默认选择1k行
    将输出转化成numpy输出，返回X, y
    '''
    for df in dflist:
        X=LinearModel_X(df)
        y=LinearModel_y(df)
        yield X, y
# 生成测试文件
minibatch_test_iterators = iter_minibatches(NCdflist)
X_test, y_test = minibatch_test_iterators.__next__()  # 得到一份测试文件
```

### SGDRegreessor的归一化

```python
from sklearn.preprocessing import StandardScaler
X_all=LinearModel_X(alldata)
y_all=LinearModel_y(alldata)
SGD_XScaler = StandardScaler().fit(X_all)
SGD_YScaler = StandardScaler().fit(y_all)
```

## 增量学习

```python
from sklearn.linear_model import SGDRegressor
sgd = SGDRegressor()  # SGDClassifier的参数设置可以参考sklearn官网
minibatch_train_iterators = iter_minibatches(NCdflist)
for i, (X_train, y_train) in enumerate(minibatch_train_iterators):
    X_train=SGD_XScaler.transform(X_train)
    y_train=SGD_YScaler.transform(y_train)
    sgd.partial_fit(X_train, y_train)
    print("{} time".format(i))  # 当前次数
    print("{} score".format(sgd.score(X_test, y_test)))  # 在测试集上看效果
```

```python
#准备测试数据
df=testlist[0]
test_X=LinearModel_X(df)
test_y=LinearModel_y(df)
test_X=SGD_XScaler.transform(test_X)
test_y=SGD_YScaler.transform(test_y)
pred_y=sgd.predict(test_X)
```
