```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
alldata=pd.DataFrame()
AllDataLength=419
NClist=[]
for i in range(AllDataLength):
    df=pd.read_csv(r"C:\Users\14020\Desktop\NCBDC 2019\Data\NoNormaliezdData\continueTimedataNC"+'\\'+str(i)+".csv").iloc[:,1:]
    alldata=pd.concat([alldata,df])
    NClist.append(df)
```

```python
InputFeatures=[ 'total_voltage', 'total_current','soc','temp_max','temp_min', 'motor_voltage', 'motor_current', 'total_P', 'motor_P','tempMAXMINdiff','SOCgap']
OtherFeatures=['speed','mileage', 'milediff','speeddiff' ,'milegap']
```

```python
scaler = StandardScaler().fit(alldata.loc[:,InputFeatures])   
```

```python
#clearTestData正则化
path = lambda car,number:r"C:\Users\14020\Desktop\NCBDC 2019\Data\Test\Test\Vehicle No.0"+str(car)+"_Fragment No."+str(number)+'.csv'
Testlist=[]
for i in range(15,30):
    for j in range(100,120):
        df=pd.read_csv(path(i,j))
        index=df.loc[df.soc==0].index
        df=df.drop(index)
        df['total_P']=df.total_voltage*df.total_current
        df['motor_P']=df.motor_current*df.motor_voltage
        df['tempMAXMINdiff']=df.temp_max-df.temp_min
        df['SOCgap']=df.soc-df.loc[0,:].soc
        DF=pd.DataFrame(scaler.transform(df.loc[:,InputFeatures]),columns=InputFeatures)
        DF['mileage']=df.mileage
        DF['time']=df.time
        DF.to_csv(r"C:\Users\14020\Desktop\NCBDC 2019\Data\Test\NormalizedClearTestData\car"+str(i)+'frag'+str(j)+'.csv')
```
